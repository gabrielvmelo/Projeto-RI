```python
import math
import numpy as np

def norm (document):
        size = [x**2 for x in document]
        size = math.sqrt(sum(size))
        if (size != 0):
            normalized = [x/size for x in document]
        else: normalized = document
        
        return normalized

def cossine(document, query):
        document = norm(document)
        query = norm(query)
        cos = np.dot(document, query)
        return cos
```

```python
def build_vec(query, document=False, boolean=False):
        result = set()
        for item in query:
            for q in query[item]:
                #print(q)
                #print(query[item])
                if not document or boolean:
                    result.add(q+'.'+item)
                else:
                    result.add(((q+'.'+item), q))
        return list(result)
```

```python
def build_vec2(query, document=False, boolean=False):
        result = []
        for item in query:
            for q in query[item]:
                if not document or boolean:
                    result.append(q+'.'+item)
                else:
                    result.append(((q+'.'+item), query[item][q]))
        return result
```

Consulta sem peso

```python
def query_weightless(query):
        query = build_vec(query)
        r = [1 for x in query]
        return r
```

Consulta com pesos

```python
def query_weight(query):
        r = []
        q2 = build_vec2(query)
        query = build_vec(query)
        for term in query:
            r.append(q2.count(term))
        return r
```

```python
zone_score = {'storyline': 0.44, "premiere_date": 0.65, 'title': 1, 'network': 0.3}

def document_weightless(query, document, zone_score_bool=False):
        query_vec = query_weightless(query)
        query = build_vec(query)
        rank = []
        
        for doc in document:
            cur_doc = build_vec(document[doc], document=True, boolean=True)
            
            vec = []
            for term in query:
                if not zone_score_bool:
                    if(term in cur_doc): vec.append(1)
                    else: vec.append(0)
                else:
                    t = term.split('.')
                    if term in cur_doc: vec.append(zone_score[t[1]])
                    else: vec.append(0)
            cos = cossine(vec, query_vec)
            rank.append((doc, cos))
        return rank
```

```python
def doc_freq(doc):
    r = {}
    
    for att in doc:
        tf_list = {}
        word_list = doc[att]
        word_unique = list(set(word_list))
        #print(att)
        #print(word_list)
        for word in word_unique:
            #print(word_list)
            tf_list[word] = word_list.count(word)
        
        r[att] = tf_list
        
    return r
        
doc_freq(inverted_file["7449632"])
```

```python
    def document_weight(query, document, BM25=False, b=0.75, k1=1.25, zone_score=False):
        result = {}
        rank = []
        idf = {}
        n_doc = len(document)
        mean_docs = media #quantidade total de palavras, sobre o total de documentos
        query_vec = query_weight(query)
        query = build_vec(query)
        
        for doc in document:
            document[doc] = doc_freq(document[doc])

        for doc in document:
            tf = {}
            #print(document[doc])
            cur_doc = build_vec(document[doc], document=True)
            size_doc = len(build_vec(document[doc]))
            
            for term in query:
                vec = 0
                has = [True for x in cur_doc if x[0] == term]
                if(len(has) > 0):
                    q = term.split('.')
                    if not zone_score:
                        vec = document[doc][q[1]][q[0]]
                    else: vec = vec*zone_score[q[1]]
                    if(BM25):
                        num = vec*(k1 + 1)
                        freq = b*size_doc/mean_docs
                        den = vec + k1*(1-b + freq)
                        vec = num/den
                    if(term in idf.keys()): idf[term] += 1
                    else: idf.update({term: 1})
                else:
                    vec = 0
                    if(term in idf.keys()): idf[term] += 0
                    else: idf.update({term: 0})
                tf.update({term: vec})
            result.update({doc: tf})
        for x in result:
            r = []
            for term in query:
                if idf[term] != 0:
                    if idf[term] == 0 or n_doc/idf[term] == 0:
                        tfidf = 0
                    else:
                        tfidf = result[x][term]*math.log10(n_doc/idf[term])
                else:
                    tfidf = 0
                r.append(tfidf)
            cos = cossine(r, query_vec)
            rank.append((int(x), cos))
        return rank
```

```python
def kendal(r1, r2):
    rank1 = r1
    rank2 = r2
    p = {}
    for i in range(0,len(rank1)):
        for j in range(i+1, len(rank1)):
            if p.get((rank1[i],rank1[j]),-1) == -1:
                p[(rank1[i],rank1[j])] = 0
            p[(rank1[i],rank1[j])] += 1
            if p.get((rank2[i],rank2[j]),-1) == -1:
                p[(rank2[i],rank2[j])] = 0
            p[(rank2[i],rank2[j])] += 1
    dis = 0.0
    for k in p.keys():
        if(p[k]==1):
            dis += 1
    dis *= 2.0
    k = len(rank1)
    den = k*(k-1)
    ans = 1 - (dis/den)
    return ans

kendal(r1, r2)
```
